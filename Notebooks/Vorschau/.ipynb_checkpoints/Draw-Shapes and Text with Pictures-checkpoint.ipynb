```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cv2.__version__
```

```python
path = "../../Pictures/dogs_vs_cats/PetImages/Dog/4.jpg"
```

```python
img = cv2.imread(path, 0)

cv2.imshow('image',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img,cmap="gray")
plt.show()
```

## Line

```python
img = cv2.imread(path, 1)

#line(image, point1, point2, color(BGR), thickness)
img = cv2.line(img, (0,0), (255,255), (0,255,0), 10) #Color in BGR-Format
#img = cv2.arrowedLine(img, (0,0), (255,255), (0,255,0), 10) #Color in BGR-Format

cv2.imshow('image',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Rectangle

```python
img = cv2.imread(path, 1)

#rectangle(image, TopLeftCoordinate, LowerRightCoordinate, color(BGR), thickness, lineType, shift)
img = cv2.rectangle(img, (120,150), (170,200), (0,0,255), 10) ## erst Breite dann Höhe | Achtung indexing in numpy ist erst height dann width !!!
#img = cv2.rectangle(img, (120,150), (170,200), (0,0,255), cv2.FILLED) ## ausgefülltes Rechteck oder -1
cv2.imshow('image',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Circle

```python
img = cv2.imread(path, 1)


img = cv2.circle(img, (150,170), 20, (0,0,255), -1) ## thickness=-1 für Ausgefüllt

cv2.imshow('image',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Text

```python
img = cv2.imread(path, 1)

font = cv2.FONT_HERSHEY_SIMPLEX
# putText(image, Text, Startpunkt des Textes, FontObjekt, Größe, Farbe, Thickness, LineType)
img = cv2.putText(img, 'Mein Hund Leo', (10,260), font, 1, (0,0,255), 2, cv2.LINE_AA)

cv2.imshow('image',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
