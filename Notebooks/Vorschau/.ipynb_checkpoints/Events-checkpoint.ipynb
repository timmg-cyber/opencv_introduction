```python
def click_event(event, x, y, flags, param): #Callback function für Eventlistener
    if event == cv2.EVENT_LBUTTONDOWN:
        font = cv2.FONT_HERSHEY_SIMPLEX
        text = "X(B):"+ str(x)+" , Y(H):"+str(y) 
        cv2.putText(img, text, (x,y), font, .5, (0,0,255), 2)
        cv2.imshow('image',img)
    if event == cv2.EVENT_RBUTTONDOWN:
        blue = img[y,x,0] #Höhe,Breite,blauer Channel (BGR)
        green = img[y,x,1]
        red = img[y,x,2]

        font = cv2.FONT_HERSHEY_SIMPLEX
        text = "("+ str(blue)+","+ str(green)+","+str(red)+")"
        cv2.putText(img, text, (x,y), font, .5, (0,255,255), 2)
        cv2.imshow('image',img)
    
#Nachvollziehen der Höhe & Breite
#img = cv2.imread(path)
#img = cv2.resize(img, (img.shape[1]*2,img.shape[0]*2),interpolation=cv2.INTER_AREA)

plt.imshow(img)
plt.show()

cv2.imshow('image', img)

cv2.setMouseCallback('image', click_event) #WindowName muss überall gleich sein

cv2.waitKey(0)
cv2.destroyAllWindows()
```
