```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path1= "../../Pictures/data/messi5.jpg"
path2= "../../Pictures/data/opencv-logo.png"
```

```python
#Einlesen und Anzeigen der Bilder
img = cv2.imread(path1)
img2 = cv2.imread(path2)

cv2.imshow('image',img)
cv2.imshow('image2',img2)
           
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Mergen der Bilder mit .add-Methode

```python
img = cv2.imread(path1)
img2 = cv2.imread(path2)

#Resizing, da beide Bilder die gleiche Größe haben müssen
img = cv2.resize(img, (512,512))
img2 = cv2.resize(img2, (512,512))

added_img = cv2.add(img, img2) 

cv2.imshow('image',added_img)
           
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## mit numpy

```python
img = cv2.imread(path1)
img2 = cv2.imread(path2)

#Resizing, da beide Bilder die gleiche Größe haben müssen
img = cv2.resize(img, (512,512))
img2 = cv2.resize(img2, (512,512))

#added_img = cv2.add(img, img2)
added_img = img + img2 
## Modulo-Operation wird durchgeführt und die Restwerte übertragen --> besser bei OpenCV-Funktion bleiben

cv2.imshow('image',added_img)
           
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Mergen der Bilder mit .addWeighted-Methode

```python
img = cv2.imread(path1)
img2 = cv2.imread(path2)

#Resizing, da beide Bilder die gleiche Größe haben sollen (muss nicht immer sein)
img = cv2.resize(img, (512,512))
img2 = cv2.resize(img2, (512,512))

added_img = cv2.addWeighted(img, .8, img2, .2,0) # alpha: Weighting von erstem Bild, Beta: Weighting von zweitem Bild, Gamma: Skalar welches draufgerechnet wird

cv2.imshow('image',added_img)
           
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Mergen von Bildern bei spezieller ROI (Region of Interest)

```python
img = cv2.imread(path1)
img2 = cv2.imread(path2)
```

```python
img.shape
```

```python
plt.imshow(img) ##Farbchannel erstmal egal
plt.show()
```

```python
# ca 30 bis 30+50 (80) Breite
# ca  shape[0]-50 (262) bis 312 Höhe
```

```python
roi = img[img.shape[0]-80:img.shape[0]-30,30:30+50]
```

```python
plt.imshow(roi)
plt.show()
```

```python
roi.shape
```

```python
img2 = cv2.resize(img2, (50,50), interpolation=cv2.INTER_AREA)
```

```python
##add image
merged = cv2.addWeighted(roi, .8, img2, .2,0)
```

```python
##Overwrite original image at ROI
img[img.shape[0]-80:img.shape[0]-30,30:30+50] = merged
```

```python
cv2.imshow('image', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
