```python
import cv2
import matplotlib.pyplot as plt
```

```python
path = "../../Pictures/dogs_vs_cats/PetImages/Dog/4.jpg"
```

## Untersuchen des Bildes

```python
img = cv2.imread(path, 0)

plt.imshow(img, cmap="gray")
plt.show()
```

```python
img.shape #heigt, width
```

```python
width = img.shape[1]
height = img.shape[0]
print("h:"+str(height)+" x"+ " w:"+str(width))
print(f"h:{height} x w:{width}")
```

## Resizing

```python
#Bild um 25% Vergrößern

new_width = int(width*1.25)
new_height = int(height*1.25)

dimensions = (new_width,new_height) # Erst breite dann Höhe in OpenCV

##Resizing mit aspect ratio
img_reshaped = cv2.resize(img, dimensions, interpolation=cv2.INTER_AREA)

##Resizing ohne aspect ration
img_reshaped2 = cv2.resize(img, (200,100), interpolation=cv2.INTER_AREA)
img_reshaped.shape
```

```python
plt.imshow(img_reshaped2, cmap="gray")
plt.show()
```

```python
plt.imshow(img_reshaped, cmap="gray")
plt.show()
```
