## Erstellen von Bildern mit Einfarbigem Hintergrund

```python
#Erstellung eines schwarzen Bildes (height,width,channels)
img = np.zeros([512,512,3], np.uint8) # uint-8 ist ein Datentyp (man k√∂nnte auch int32 nehmen)

#√Ñndern der Farbe
img[:] = 0,0,255 ##BGR 

#Teil des Bildes wei√ü f√§rben
img[0:50,0:512] = 255,255,255     # H√§heAnfang-H√§heEnde, BreiteAnfang-BreiteEnde

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
